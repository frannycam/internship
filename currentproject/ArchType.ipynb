```python
import pandas as pd 
import numpy as np
import math
import xlrd
import csv

df = pd.read_csv('original.csv')

zScoreR=[]
zScoreL=[]
percentileR=[]
percentileL=[]
corrected_archTypeR=[]
corrected_archTypeL=[]
sLength = len(df['Gender'])

def main():
    SIZE=df['ShoeSize_us']  
    GENDER = df['Gender']
    for i in range(len(GENDER)):
        for j in range(len(SIZE)):
            gender=GENDER[i]
            size=SIZE[j]
            arch_meanStdDev(gender,size, i)
    df.assign(zScoreR=zScoreR.values).head()
    df.assign(zScoreR=zScoreL.values).head()
    df.assign(zScoreR=percentileR.values).head()
    df.assign(zScoreR=percentileL.values).head()
    df.assign(zScoreR=zScoreR.values).head()
    df.assign(zScoreR=zScoreR.values).head()
    

#     df.loc[:,'zScoreR'] = np.Series(np.random.randn(sLength), index=df.index)
#     df.loc[:,'zScoreL'] = np.Series(np.random.randn(sLength), index=df.index)
#     df.loc[:,'percentileR'] = np.Series(np.random.randn(sLength), index=df.index)
#     df.loc[:,'percentileL'] = np.Series(np.random.randn(sLength), index=df.index)
#     df.loc[:,'corrected_archTypeR'] = np.Series(np.random.randn(sLength), index=df.index)
#     df.loc[:,'corrected_archTypeL'] = np.Series(np.random.randn(sLength), index=df.index)
# #     df = df.assign(zScore=p.Series(np.random.randn(sLength)).values)

            

def arch_meanStdDev(gender,size,i):
    arch_data_R =[] #new array
    arch_data_L =[] 
    RightArch = df['RightArchHeightMm']
    LeftArch = df['LeftArchHeightMm']
    US_Sizes = [5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10,10.5,11,11.5,12,12.5,13,13.5,14]
    Gender_and_Size =df.set_index(['Gender', 'ShoeSize_us'])
    if gender =="Female" or gender=="Male":
        if size in US_Sizes:
            onlysize_data = Gender_and_Size.loc[gender, size] 
            arch_data_R = onlysize_data['RightArchHeightMm'] 
            arch_data_L = onlysize_data['LeftArchHeightMm']
            total_data = arch_data_L + arch_data_R 
            means=np.mean(total_data)
            std_dev=np.std(total_data)  
            x=RightArch[i]
            zScore=(x-means)/std_dev
            zScoreR.append(zScore)
            percentile=(.5 * (math.erf(zScore / 2 ** .5) + 1))*100
            percentileR.append(percentile)
            archtype=arch_Type(percentile)
            corrected_archTypeR.append(archtype)
            x=arch_data_L[i]
            zScore=(x-means)/std_dev
            zScoreL.append(zScore)
            percentile=(.5 * (math.erf(zScore / 2 ** .5) + 1))*100
            percentileL.append(percentile)
            archtype=arch_Type(percentile)
            corrected_archTypeL.append(archtype)
            
def arch_Type(percentile): 
    if percentile<20.0:
        return 'low'
    elif percentile>20.0 and percentile<80.00:
        return 'medium'
    elif percentile>80.00:
        return'high'            

main()

# if 1>0:
#     main()
#     df = df.assign(zScoreR=zScoreR.values)
```
